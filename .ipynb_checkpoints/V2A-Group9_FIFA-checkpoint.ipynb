## prepareer Spelers-dataset <a id='Spelers_prepareer'></a>

De spelers-dataset bevat fouten die wij in dit gedeelte zullen oplossen.

Verbeterpunten die zijn doorgevoerd op de spelers dataset zijn:

1. Irrelevante kolommen filteren.
2. Alle rijen met lege kolommen verwijderen.
3. Alle letters filteren uit Weight.
4. Alle apostrofs filteren uit Height.
5. Alle waardes van Value omzetten naar volledig uitgeschreven getallen.
6. Spellingscontrole op alle clubnamen.
7. spelers met spelerswaarde nul verwijderen.

We kunnen zien dat 'Weight', 'Height' en 'Value' dtypes van object hebben wat inhoud dat ze een string zijn. Dit verwachten we niet omdat we zouden denken dat het numerieke waardes zijn, dit zullen we corrigeren in het onderstaande blok.

```python
players.dtypes
```

```python
# verwijder alle rijen die lege kolommen bevatten.
players = players.dropna()

# We halen de lbs aan het einde weg en zetten we het om in een float
players['Weight'] = players['Weight'].str[:-3].astype(float)

# We halen de ' weg en vervangen het met . en vervolgens zetten we het om in een float
players['Height'] = players['Height'].str.replace("'", ".").astype(float)

# We halen de  en . weg, vervolgens vervangen we M met 6 nullen en K met 3 nullen en zetten het om in een int
players['Value'] = players['Value'].str.replace('', '').str.replace('.', '')
players['Value'] = players['Value'].str.replace('M', '000000')
players['Value'] = players['Value'].str.replace('K', '000').astype(int)

# We halen de  en . weg, vervolgens vervangen we M met 6 nullen en K met 3 nullen en zetten het om in een int
players['Wage'] = players['Wage'].str.replace('', '').str.replace('.', '')
players['Wage'] = players['Wage'].str.replace('M', '000000')
players['Wage'] = players['Wage'].str.replace('K', '000').astype(int)

# spellingscontrole op alle clubnamen
players['Club'] = players['Club'].str.replace('1. ', '')
players.dtypes
```

## Spider plot <a id='Spider_plot'></a>

```python
# Alleen relevante data selecteren en de beste en slechtste speler
player_skills = ['Name', 'Agility', 'BallControl', 'Dribbling', 'Finishing','Stamina']

# Alle spelers met de releveante data ophalen
all_players = players[player_skills]
all_players = all_players.set_index('Name')

# Een lijst met namen van de 5 beste en 5 slechtse spelers
best_players = list(all_players.head().index.values)
worst_players = list(all_players.tail().index.values)

# 2 Dropdown list widgets om een speler te kunnen selecteren
player1_widget = widgets.Dropdown(options = best_players, value = best_players[0], description='Player 1:')
player2_widget = widgets.Dropdown(options = worst_players, value = worst_players[-1], description='Player 2:')

# Functie om gegeven 2 namen de spelers in een spiderplot te kunnen zetten
def spiderplot_players(p1, p2):
    # Groote van de plot bepalen
    plt.figure(figsize=(20, 20))
    
    # De kollomen van de spelers ophalen
    comparison_players = all_players.loc[[p1, p2]] 

    for name, skills in comparison_players.iterrows():
        categories = skills.index.values

        # Alle punten om te tekenen
        points = list(skills.values)
        points += points[:1]

        # Alle hoeken berekenen
        angles = [side / float(len(categories)) * 2 * np.pi for side in range(len(categories))]
        angles += angles[:1]

        # De spider plot aanmaken
        sp = plt.subplot(3, 3, 1, polar=True)

        # De categorieen weergeven
        plt.xticks(angles, categories)

        # De markers weergeven
        plt.yticks([25, 50, 75], ['25', '50', '75'], color='grey')
        plt.ylim(0,100)

        # Alles plotten en opvullen
        sp.plot(angles, points)
        sp.fill(angles, points, alpha=0.1)

    # De titel plotten
    names = comparison_players.index.values
    plt.title(f'{names[0]} vs {names[1]}', y=1.1)
    
interactive(spiderplot_players, p1=player1_widget, p2=player2_widget)
```

## Onderzoeksvraag 2 <a id='Onderzoeksvraag_2'></a>

__Kan je op basis van de gemiddelde waardes van een team voorspellen welke van de twee teams zullen winnen?__

door middel van supervised machine learning gaan we berekenen hoe groot de kans is dat we een wedstrijduitslag juist  voorspellen. Hierbij categoriseren we alle spellers per club. waarna we de 14 beste spelers selecteren op basis van hen kwaliteit.

In dit voorbeeld breiden we de Nederlandse competitie dataset uit.

```python
def functie(competitie, feature, average, homeColumn, awayColumn):
    # spelers sorteren per club
    ordered = players.groupby('Club')[feature].apply(list).reset_index()

    # gemiddelde waarde berekenen van spelers kwaliteit.
    avg_feature = []
    for row in ordered[feature]:
        a = row[:14]
        a.sort(reverse = True)
        a = round(sum(a) / len(a),3)
        avg_feature.append(a)
    ordered[average] = avg_feature
    ordered = ordered.drop(columns=[feature])

    # wedstrijduitslag voorspellen op basis van spelers kwaliteit
    ht_lijst, at_lijst = [], []
    for i in range(len(competitie)):
        for j in range(len(ordered)):
            if competitie['HomeTeam'][i] == ordered['Club'][j]:
                ht_lijst.append(ordered[average][j])
            if competitie['AwayTeam'][i] == ordered['Club'][j]:
                at_lijst.append(ordered[average][j])
    
    # dataset uitbreiden met onderzoeksresultaten.
    competitie[homeColumn] = ht_lijst
    competitie[awayColumn] = at_lijst

    competitie[feature] = np.where(competitie[homeColumn] > competitie[awayColumn], 0, 2)
    competitie[feature] = np.where(competitie[homeColumn] == competitie[awayColumn], 1, competitie[feature])

# bepalen welke team heeft gewonnen.
nl['WT'] = np.where(nl['FTHG'] > nl['FTAG'], 0, 2)
nl['WT'] = np.where(nl['FTHG'] == nl['FTAG'], 1, nl['WT'] )

functie(nl, 'Stamina', 'avg_Stamina', 'Stamina_HT', 'Stamina_AT')
functie(nl, 'Overall', 'avg_Overall', 'Overall_HT', 'Overall_AT')
functie(nl, 'Vision', 'avg_Vision', 'Vision_HT', 'Vision_AT')

nl.describe()
```

```python
nl.sample(15)
```

Het uitbreiden van de dataset stelt ons in staat om de informatie vanuit een andere hoek te kunnen bekijken. Zo zie je dat de betere club niet altijd wint.

Laten we de informatie overzichtelijk presenteren door middel van graphishe anlyse. Zo ook zullen we met behulp van DecisionTree een poging te wagen om de wedstrijden zo goed mogelijk te voorspellen.

```python
model = DecisionTreeClassifier()

# bepaal features & target variabele
feat_names = ['Stamina', 'Overall', 'Vision']
feats = np.array(nl[feat_names])
cat = nl['WT']

feats_train, feats_test, cat_train, cat_test = train_test_split(feats, cat, random_state=10)
print('{} elementen in de trainset en {} elementen in de testset'.format(len(cat_train), len(feats_test)))
```

```python
model.fit(feats_train, cat_train)
# M.b.v. de score wordt de accuracy berekend: het percentage voorspellingen dat correct
model.score(feats_test, cat_test)
```

```python
# Alternatieve manier om de voorspellende waarde te berekenen is door het model de klasse te voorspellen voor
# de testset van de features en deze te vergelijken met de testset van de klassen.
cat_model = model.predict(feats_test)
accuracy_score(cat_test, cat_model)
```

Voorafgaand zijn er berekeningen utigevoerd met als doel een zo hoog mogelijk slagingspercentage. Zo blijkt dat binnen de Nederlandse competitie, spelers met een goed ontwikkeld uithoudingsvermogen, inzicht, een groter verschil kunnen uitmaken op de wedstrijduitslag.

```python
# Het is ook interessant om te kijken welke voorspellingen goed gaan en welke niet gaan. 
# Je ziet dat vooral de NY-huizen niet goed voorspeld worden. Dus het is zaak om onderscheidende features toe te voegen.
cm = confusion_matrix(cat_test, cat_model)

# Maak van de array een pandas dataframe om te visualiseren
df_cm = pd.DataFrame(cm)

fig, ax = plt.subplots(figsize=(6, 5), dpi=100)

ax = sns.heatmap(df_cm, annot=True, cmap='OrRd', fmt='d')
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)

ax.set_xlabel('voorspelde waarde')
ax.set_ylabel('echte waarde')

plt.show()
```

Hierbij zullen wij clustering gaan gebruiken om mogelijke patronen te ontdekken. 
Om die clusteringen te visualiseren gaan we Kmeans en GMM gebruiken en kijken welke van de 2 we het beste kunnen gebruiken.

Om makkelijk modelen te maken maken we functies om dit te doen.

Functies voor maken van kMeans en GMM plot.

```python
# kMeans model func
def plot_kMeans(data, number_clusters, Xas, Yas):
    
    #Make kMeans model.
    model_kMeans = KMeans()
    
    #chose data.
    kMeans = data
    
    #set number of clusters.
    model_kMeans.set_params(n_clusters = number_clusters, random_state = 0)
    
    #fit.
    model_kMeans.fit(kMeans)
    
    #show data.
    prediction_kMeans = model_kMeans.fit_predict(kMeans)
    
    #make 2d plot.
    plt.scatter(kMeans[Xas],kMeans[Yas],marker='o', c=prediction_kMeans)
    plt.scatter(model_kMeans.cluster_centers_[:,0],model_kMeans.cluster_centers_[:,1],marker='x', s=40)
    plt.gcf().set_size_inches((10, 10))  
    
    return plt

# GMM model func
def plot_GMM(data, number_componets, Xas, Yas):

    #Make kMeans model.
    model_GMM = GaussianMixture()
    
    #chose data.
    data_GMM = data
    
    #set number of clusters.
    model_GMM.set_params(n_components = number_componets, random_state = 0)
    
    #fit.
    model_GMM.fit(data_GMM)
    
    #show data.
    prediction_GMM = model_GMM.fit_predict(data_GMM)
    
    #make 2d plot.
    plt.scatter(data_GMM[Xas],data_GMM[Yas],marker='o', c=prediction_GMM)
    plt.scatter(model_GMM.means_[:,0],model_GMM.means_[:,1],marker='x', s=40)
    plt.gcf().set_size_inches((10, 10))   
    
    return plt
```

```python
data = []
max_n = 20

#for i in range(1,len(dataset)):
for i in range(1, max_n):
    model_kMeans = KMeans(n_clusters=i, random_state=0)
    prediction_kMeans = model_kMeans.fit_predict(nor_data1)
    data.append([i, model_kMeans.score(nor_data1)])

df = pd.DataFrame(data, columns=['k', 'Score'])

fig = plt.figure(figsize=(5,5), dpi=150)

# object axes: de grafiekbasis
ax = plt.axes()

ax.set(xlim=(0,max_n),
       xlabel='k',
       ylabel='Score')      

ax.xaxis.set_major_locator(plt.MaxNLocator(20))
ax.ticklabel_format(useOffset=False)
ax.plot(df['k'], df['Score'], '-o')
```

```python
#make clusters.
prediction_kMeans = model_kMeans.fit_predict(nor_kMeans_data)
```

Om de verbanden tussen Height, Weight, Age en Value te leggen moeten we de clusters met elkaar vergelijken.
Dit gaan we doen door voor elke cluster de mediaan te pakken van Height, Weight, Age en Value.
We kiezen voor de mediaan omdat die het best aangeeft waar de cluster zich bevind in de data.
De medianen zullen we dan in een barplot laten zijn om de clusters te vergelijken.

clusters bij de data toevoegen.

```python
# add clusters to data.
nor_kMeans_data['Cluster_Number'] = prediction_kMeans
```

Clusters scheiden.

```python
# get clusters
cluster1 = nor_kMeans_data.loc[nor_kMeans_data['Cluster_Number'] == 0]

cluster2 = nor_kMeans_data.loc[nor_kMeans_data['Cluster_Number'] == 1]

cluster3 = nor_kMeans_data.loc[nor_kMeans_data['Cluster_Number'] == 2]

cluster4 = nor_kMeans_data.loc[nor_kMeans_data['Cluster_Number'] == 3]

cluster5 = nor_kMeans_data.loc[nor_kMeans_data['Cluster_Number'] == 4]
```

Medianen berekenen van alle clusters en ze bij elkaar in een lijst zetten.

```python
# Height mediaan van alle clusters
Cluster_Height = [cluster1.loc[:,'Height'].median(), cluster2.loc[:,'Height'].median(),
                  cluster3.loc[:,'Height'].median(), cluster4.loc[:,'Height'].median(),
                  cluster5.loc[:,'Height'].median()]

# Weight mediaan van alle clusters
Cluster_Weight = [cluster1.loc[:,'Weight'].median(), cluster2.loc[:,'Weight'].median(),
                  cluster3.loc[:,'Weight'].median(), cluster4.loc[:,'Weight'].median(),
                  cluster5.loc[:,'Weight'].median()]

# Age mediaan van alle clusters
Cluster_Age = [cluster1.loc[:,'Age'].median(), cluster2.loc[:,'Age'].median(),
               cluster3.loc[:,'Age'].median(), cluster4.loc[:,'Age'].median(),
               cluster5.loc[:,'Age'].median()]

# Value mediaan van alle clusters
Cluster_Value = [cluster1.loc[:,'Value'].median(), cluster2.loc[:,'Value'].median(),
                 cluster3.loc[:,'Value'].median(), cluster4.loc[:,'Value'].median(),
                 cluster5.loc[:,'Value'].median()]

# set index
index = ['cluster1', 'cluster2', 'cluster3', 'cluster4', 'cluster5']
```

Plot maken van de bar diagram.

```python
# bar plot data aanwijzen
pb = pd.DataFrame({'Height': Cluster_Height,
                   'Weight': Cluster_Weight,
                   'Age': Cluster_Age,
                   'Value': Cluster_Value}, index=index)

# set plot size
plt.rcParams["figure.figsize"] = [15, 10]

# show plot
pb.plot.bar()
```

Wat gelijk opvalt is de cluster2 een hele hoge value heeft en de andere clusters gelijkwaardige value maar wel in de min. Dat komt niet omdat hun value zo laag is maar omdat er een paar uitschieters bij value zijn.

cluster1 zijn spelers met: normale height, hele lage weight, lage age en een normale value.

cluster2 zijn spelers met: normale height, weight en een gemiddelde age waarvan de value heel hoog is.

cluster3 zijn spelers met: hoge height, weight en hele hoge age met een normale value.

cluster4 zijn spelers met: hoge weight, hoge weight, lage age en normale value.

cluster5 zijn spelers met: hele lage height, lage weight, lage age en normale value.

Uit de resultaten zie we dat er een verband is tussen de lengte en gewicht van een speler. Als een speler een een hoge lengte heeft heeft die waarschijnlijk ook een hoog gewicht en anders om ook. De enige uitzodering zien we bij cluster1 waarbij er een normale lengte is en een heel laag gewicht.

We kunnen ook een verband zien in de leeftijd en value. de duurste spelers zijn de gene die niet te jong zijn of te oud. Dit zie je goed bij cluster1 waarbij de leeftijd normaal is en de value heel hoog. Dit zijn waarschijnlijk de jaren waar een speler op zijn best is of net daarvoor.

### conclusie

Er zijn geen verbanden tussen alle vier height, weight, age en value van spelers. 
Dit komt omdat de data niet goed in clusters verdeeld kon worden of, door de simpele reden dat er geen verbanden zijn.
Dat valt te verklaren doordat er bij voetbal geen duidelijke voordelen of nadelen zijn bij een bepaald gewicht, lengte, leeftijd.

## Onderzoeksvraag 4 <a id='Onderzoeksvraag_4'></a>

_Verschilt de potentie van de huidige generatie Engelse spelers sterk van spelers uit de rest van de wereld?_

```python
english_potential = players[players['Nationality'].isin(['England'])]['Potential']
world_potential = players[~players['Nationality'].isin(['England'])]['Potential']
percentage_change = ((world_potential.mean() - english_potential.mean()) / world_potential.mean()) * 100

print(f'Average Potential of all footballplayers worldwide(excluding. English men): {round(world_potential.mean(), 2)}')
print(f'Average potentialnly of all English footballplayers: {round(english_potential.mean(), 2)}')
print(f'Engeland scoort {round(percentage_change, 2)}% lager dan de rest van de wereld.')
print('worldwide average potential:', round(players['Potential'].mean(),2))
print('amount of worldwide footballplayers:', len(players))
print('amount of English footballplayers:', sum(players['Nationality'] == 'England'))
print('amount of English footballplayers above worldwide average potential:',len(players.loc[(players['Potential'] > 80) & players['Nationality'].isin(['England'])]))
```

We kunnen hier zien dat het verschil gering is en daarmee kunnen we concluderen dat de huidige generatie Engelse spelers niet sterk van spelers uit de rest van de wereld verschillen.

```python
combined = pd.concat([world_potential, english_potential], axis=1)
combined.columns = ['world_potential', 'english_potential']
bin_size = len(world_potential.unique())
combined.plot.hist(bins=bin_size, alpha=0.5)
```

We zien in deze histogram dat de vorm erg soortgelijk is en het middelpunt beide bij 70 zit, hier gaan we beneden verder op in.

```python
frequencies, edges = np.histogram(english_potential, len(english_potential))
hv.Histogram((edges, frequencies))
```

We kunnen zien dat het gemiddelde rond de 70 zit en de moduus ook.

```python
frequencies, edges = np.histogram(world_potential, len(world_potential))
hv.Histogram((edges, frequencies))
```

We kunnen zien dat er een hele mooie historgram uitkomt, het lijkt bijna op een normaalverdeling. Zoals we zien ligt ook hier het gemiddelde rond de 70.

```python
print(f'English potential modus: {english_potential.mode().values[0]}')
print(f'World potential modus: {world_potential.mode().values[0]}')
```

Voor de zekerheid vragen we beide modus op en zoals we al konden aflezen ligt de modus erg dicht bij elkaar.
